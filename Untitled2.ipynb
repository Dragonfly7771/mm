Лабораторная 3. Ивашин Е.В.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(y, t):
    k = 0.42
    return -k*y

t1=np.linspace(0, 10, 100)

solve = odeint(dydt, y0=1, t=t1)
solve = solve.flatten()


h = 0.1
t2 = np.arange(0, 10+h, h)

def euler(f, y0, t, h=0.1):
    y = []
    y.append(y0) 
    for i in range(1, len(t)):
        yn = y[i-1] + h * f(y[i-1] ,t[i-1])
        y.append(yn)
    return y

euler = euler(dydt, 1, t2, h)

%matplotlib inline

plt.plot(t1, solve, label = r"$y'=ky$")
plt.plot(t2, euler, label = r"$euler$")
plt.grid(True)
plt.legend()
plt.show()

```
